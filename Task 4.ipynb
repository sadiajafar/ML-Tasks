```python
#Single line:
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
#multiple lines:
import matplotlib.pyplot as plt
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,x,z)
plt.title('Sine & Cosine Curves')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#subplots:
import matplotlib.pyplot as plt
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)
#set first subplot as active:
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title('Sine')
#set second subplot as active:
plt.subplot(2,1,2)
plt.plot(x,z)
plt.title('Cosine')
plt.show()
```
